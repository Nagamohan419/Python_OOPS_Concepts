```python
#Instance methods, Class Methods, Static methods
class Student:
    school='Telusko'
    
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
        
    def avg(self):  #Instance method
        return(self.m1+self.m2+self.m3)/3
    
    def get_m1(self): #Get method(Accessor methods) to get/fetch the values from instance variables
        return self.m1
    
    def set_m1(self,value): # set method(mutator methods) is for settin the new values for instance variables
        self.m1=value
        
    @classmethod     #Decorator need to use to work or create class methods
    def getSchoolname(cls):   #cls is the keyword need to use to pass Student class
        return cls.school
    
    @staticmethod   #This is just a  decorator for static method, to define a static method method first we need to give static method decorator
    def info():# static method
        print("This is Student class.. in abc module") # we can write anything you want
    
s1=Student(34,47,32)
s2=Student(69,32,12)


s1.set_m1(100)
print(s1.get_m1())
print(s1.avg())
print(Student.getSchoolname())
print(Student.info()) #It will print none if nothing is returning by function
```

#### Encapsulation
Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an object’s variable can only be changed by an object’s method. Those types of variables are known as private variables.

A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.

```python
class Car:
    def __init__(self,speed,color):
        self.speed=speed
        self.color=color
        
ford=Car(200,'red')
honda=Car(250,'blue')
audi=Car(300,'black')

ford.speed='dskjfh'

print(ford.speed)
print(ford.color)
    
```

- In above example the speed of ford should be of interger type but we explicitly changed to string type so to avoid such type of changes by users we use encapsulation.

To encapsulate our code we create functions/methods

```python
class Car:
    def __init__(self,speed,color):
        self.speed=speed
        self.color=color
        
    def set_speed(self,value):
        self.speed=value
        
    def get_speed(self):
        return self.speed
        
ford=Car(200,'red')
honda=Car(250,'blue')
audi=Car(300,'black')

ford.set_speed(300)
ford.speed=400
print(ford.get_speed())
print(ford.color)
```

#### Abstraction

It is the concept of object oriented programming language that shows only essential attributes and hides uneccesary information.

Objects are the building blocks of OOP. An object contains some properties and methods. we can hide them from the outer world through acess modifiers. we can provide acess only for the requires functions and properties to thee other programs. This is the general procedure to implement abstraction in oops

```python
from abc import ABC
class ClassName(ABC):
```

 Abstraction classes are meant to be the blueprint of the other class. An abstract class can be useful when we are designing large functions. An abstract class is also helpful to provide the standard interface for different implementations of components. 
 
 An Abstract class can contain the both method normal and abstract method.
An Abstract cannot be instantiated; we cannot create objects for the abstract class.

```python
#Why Abstraction is Important?
#In Python, an abstraction is used to hide the irrelevant data/class in order to reduce the complexity.
#It also enhances the application efficiency.

from abc import ABC,abstractmethod

class Car(ABC):  #we wont pass any thing in the main class we put define our code in child class
    def mileage(self):
        pass

class Tesla(Car):
    def mileage1(self):
        print("The mileage is 30kmph")
class Suzuki(Car):
    def mileage(self):
        print("This mileage is 25kmph")
        
class Duster(Car):
    def mileage(self):
        print("This mileage is 24kmph")
        
class Renault(Car):
    def mileage(self):
        print("This mileage is 27kmph")
        
#Driver code
#c=Car()  
#c.mileage() #An Abstract cannot be instantiated; we cannot create objects for the abstract class.
t=Tesla()
t.mileage1()

s=Suzuki()
s.mileage()
        
d=Duster()
d.mileage()
```

```python
def div(a,b):
    print(a/b)
    
def smart_div(func):
    def inner(a,b):#This inner function will take same parameters by div
        if a<b:
            a,b=b,a
        return func(a,b)
    
    return inner
    
div=smart_div(div)
div(2,4)
```
